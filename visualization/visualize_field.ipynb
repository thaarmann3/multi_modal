```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
spike_height = 150
spike_width = 0.2
bounds = 2.5
dn = 1000
alpha = spike_height*0.1

# Create a grid of x and y values
x = np.linspace(-bounds, bounds, dn)
y = np.linspace(-bounds, bounds, dn)
X, Y = np.meshgrid(x, y)

# Calculate the potential field z = x^2 + y^2 + spike in the middle
# Add a Gaussian spike at the center

spike = spike_height * np.exp(-(X**2 + Y**2) / (2 * spike_width**2))

Z = alpha*X**2 + alpha*Y**2 + spike
```

```python
# Create 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# Add colorbar
plt.colorbar(surf, shrink=0.5, aspect=5)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Potential (Z)')
ax.set_title('Potential Field: z = x² + y² with spike')

plt.show()

# Also create a 2D contour plot
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=20)
plt.clabel(contour, inline=True, fontsize=8)
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Potential Field Contours: z = x² + y²')
plt.axis('equal')
plt.show()
```
